```python
import pandas as pd
import gdown
import seaborn as sns

# Load the dataset
df = pd.read_csv(r"C:\Users\user\Desktop\data\fifa_data.csv")
df=pd.DataFrame(df)
print(df.head())
```

```python

# Group by country and count the number of players
country_player_counts = df['Nationality'].value_counts()

# Extract the country with the most players
country_with_most_players = country_player_counts.idxmax()
print("\nCountry with the most numbers players are :", country_with_most_players)
```

```python
# To see the structure of the Dataframe
print(df.info())

# Basic staistical details
print(df.describe())

```

```python
import matplotlib.pyplot as plt

# Get the top 5 countries
top_countries = country_player_counts.head(5)

# Plotting
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', color='blue')
plt.title('Top 5 Countries with the Most FIFA Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#  Plot a histogram to get the salary range of the players.
plt.figure(figsize=(10, 6))

sns.histplot(df['Wage'], bins=30,color='lightgreen',edgecolor='blue', kde=True)
plt.title('Salary Range of Players')
plt.xlabel('Salary (EUR)')
plt.ylabel('Number of Players')
plt.show()

```

```python
# Function to convert height in the format "5'7" to total inches
def height_to_inches(height_str):
    try:
        feet, inches = height_str.split("'")
        inches = inches.replace('"', '').strip()  # Clean the inches part
        return int(feet) * 12 + int(inches)  # Convert feet to inches and add inches
    except Exception:
        return None

# Use the function to find the index of the tallest player
tallest_player_index = df['Height'].apply(height_to_inches).idxmax()

# Get the details of the tallest player
tallest_player = df.loc[tallest_player_index]

print(f"The tallest player is: {tallest_player['Name']} with a height of {tallest_player['Height']}")
```

```python
# 6. Which club has the most number of players?
club_counts = df['Club'].value_counts()
top_club = club_counts.idxmax()
print(f"Club with the most number of players: {top_club}")
```

```python
# 7. Which foot is most preferred by the players? Draw a bar chart for preferred foot.
preferred_foot_counts = df['Preferred Foot'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=preferred_foot_counts.index, y=preferred_foot_counts.values,color = 'red')
plt.title('Most Preferred Foot')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.show()
```
